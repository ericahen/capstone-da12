```python
all = [rates19, rates20, rates21, rates22, rates23]
all
```

```python
together = pd.concat(all)
```

```python
together
```

```python
rates = together[['year','school_name', 'grad_rate']]
rates
```

```python
rates = rates.drop(rates[rates['grad_rate'] == '*'].index)
rates
```

```python
rates = rates.drop(rates[rates['grad_rate'] == '**'].index)
rates
```

```python
rates.info()
```

```python
rates['grad_rate'] =rates['grad_rate'].astype(float)
rates
```

```python
rates = rates.reset_index()
rates
```

```python
rates.hist()
```

```python
sns.histplot(data=rates, x="grad_rate")
plt.show()
```

```python
plt.plot('grad_rate', 'year', data=rates, linestyle='-', marker='o')
plt.show()
```

```python
top10 = rates.sort_values('grad_rate', ascending=False).head(10)
top10
```

```python
plt.plot('year', 'grad_rate', data=rates, linestyle='-', marker='o')
plt.show()
```

```python
bottom = rates.sort_values('grad_rate', ascending=False).tail(5)
bottom
```

```python
bottom.info()
```

```python
ax = sns.barplot(bottom, x="school_name", y="grad_rate", estimator="sum", errorbar=None)
ax.bar_label(ax.containers[0], fontsize=10)

# Set the title
ax.set_title("Lowest Graduation Rates 2019-2023", fontsize=16)  # Adjust the title size as needed

# Set the x and y labels
ax.set_xlabel("School Name", fontsize=8)  # Change label size here
ax.set_ylabel("Graduation Rate", fontsize=8)  # Change label size here

# Optionally, adjust the tick parameters
ax.tick_params(axis='x', labelsize=8)  # Change x-tick label size
ax.tick_params(axis='y', labelsize=8)  # Change y-tick label size
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Show the plot
plt.show()
```

```python
plt.plot('school_name', 'grad_rate', data=bottom, linestyle='-', marker='o')
plt.xlabel('School Name') # x-axis name
plt.ylabel('Graduation Rate') # y-axis name
plt.title('Metro Nashville Public School Graduation Rate') # Add a title
plt.show()
```
