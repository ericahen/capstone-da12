```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="white", context="talk")
rs = np.random.RandomState(8)
```

```python
grad_rate_22_23 = pd.read_excel(r'C:\DA12\capstone\capstone-da12\data\graduation rates\2022-23_school_grad_rate_suppressed.xlsx')
grad_rate_22_23
```

```python
grad_rate_22_23.columns
```

```python
grad_rate_22_23['system_name'].value_counts()
```

```python
grad_rate_23 = grad_rate_22_23[grad_rate_22_23['system_name']=='Davidson County']
grad_rate_23
```

```python
rates = grad_rate_23[grad_rate_23['student_group']=='All Students']
rates
```

```python
rates = rates[['school_name', 'grad_rate_state']]
rates
```

```python
rates.info()
```

```python
rates = rates.drop(rates[rates['grad_rate_state'] == '*'].index)
rates
```

```python
rates = rates.drop(rates[rates['grad_rate_state'] == '**'].index)
rates
```

```python
rates['grad_rate_state'] =rates['grad_rate_state'].astype(float)
rates
```

```python
rates.info()
```

```python
top5 = rates.sort_values('grad_rate_state', ascending=False).head(5)
top5
```

```python
top5.plot.barh( x='school_name', y='grad_rate_state')
```

```python
color = ['lightblue', 'blue', 'purple', 'red', 'black']

# Create bars
fig, ax = plt.subplots()
ax.barh(top5['school_name'], top5['grad_rate_state'],
       color=color)

# Show graph
plt.show()
```

```python
ax = sns.barplot(top5, x="school_name", y="grad_rate_state", estimator="sum", errorbar=None)
ax.bar_label(ax.containers[0], fontsize=10)

# Set the title
ax.set_title("Graduation Rates by School", fontsize=16)  # Adjust the title size as needed

# Set the x and y labels
ax.set_xlabel("School Name", fontsize=8)  # Change label size here
ax.set_ylabel("Graduation Rate", fontsize=8)  # Change label size here

# Optionally, adjust the tick parameters
ax.tick_params(axis='x', labelsize=8)  # Change x-tick label size
ax.tick_params(axis='y', labelsize=8)  # Change y-tick label size
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Show the plot
plt.show()
```
