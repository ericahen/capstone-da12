```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
grad_rate_21_22 = pd.read_excel(r'C:\DA12\capstone\capstone-da12\data\graduation rates\2021-22_school_grad_rate_suppressed.xlsx')
grad_rate_21_22
```

```python
grad_rate_22 = grad_rate_21_22[grad_rate_21_22['system_name']=='Davidson County']
grad_rate_22
```

```python
grad_rate_22['school_name'].unique()
```

```python
rates = grad_rate_22[grad_rate_22['student_group']=='All Students']
rates
```

```python
rates = rates[['school_name', 'grad_rate']]
rates
```

```python
rates.info()
```

```python
rates = rates.drop(rates[rates['grad_rate'] == '*'].index)
rates
```

```python
rates = rates.drop(rates[rates['grad_rate'] == '**'].index)
rates
```

```python
rates['grad_rate'] =rates['grad_rate'].astype(float)
rates
```

```python
rates.info()
```

```python
top5 = rates.sort_values('grad_rate', ascending=False).head(5)
top5
```

```python
top5.plot.barh( x='school_name', y='grad_rate')
```

```python
bottom = rates.sort_values('grad_rate', ascending=False).tail(5)
bottom
```

```python
sns.set_theme(style="white", context="talk")
rs = np.random.RandomState(8)
```

```python
ax = sns.barplot(bottom, x="school_name", y="grad_rate", estimator="sum", errorbar=None)
ax.bar_label(ax.containers[0], fontsize=10)

# Set the title
ax.set_title("Graduation Rates by School 2022", fontsize=16)  # Adjust the title size as needed

# Set the x and y labels
ax.set_xlabel("School Name", fontsize=8)  # Change label size here
ax.set_ylabel("Graduation Rate", fontsize=8)  # Change label size here

# Optionally, adjust the tick parameters
ax.tick_params(axis='x', labelsize=8)  # Change x-tick label size
ax.tick_params(axis='y', labelsize=8)  # Change y-tick label size
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Show the plot
plt.show()
```
